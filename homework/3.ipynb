```python
#homework3
#1.编写 Python程序,判断输入a是否为质数
a=int(input("请输入一个需要判断的数："))
if a<2:
    print(f"{a}不是质数")
else:
    is_prime=True
    for i in range(2,int(a**0.5)+1):
        if a%i==0:
            is_prime=False
            break
    if is_prime:
        print(f"{a}是质数")
    else:
        print(f"{a}不是质数")
```

```python
#2.用Python实现插入排序该算法，不允许直接调用插入排序函数，需要手动实现其逻辑过程。
def insertion_sort(arr):
    n=len(arr)
    for i in range(1,n):
        key=arr[i]
        j=i-1
        while j>=0 and arr[j]>key:
            arr[j+1]=arr[j]
            j-=1
            arr[j+1]=key
    return arr

s=input("请输入一组以空格分隔的整数：")
arr=[int(item) for item in s.split()]
print("输入数组：",arr)
print("插入排序之后的数组：",insertion_sort(arr))
```

```python
#3.用Python实现选择排序该算法，不允许直接调用选择排序函数，需要手动实现其逻辑过程。
def select_sort(arr):  
    n = len(arr)  
    for i in range(n):  
        min_index = i  
        for j in range(i+1, n):  
            if arr[j] < arr[min_index]:  
                min_index = j  
        if min_index != i:  
            arr[i], arr[min_index] = arr[min_index], arr[i]  
    return arr 

s=input("请输入一组以空格分隔的整数：")
arr=[int(item) for item in s.split()]
print("输入数组：",arr)
print("选择排序之后的数组：",insertion_sort(arr))
```

```python
#4.用Python实现快速排序该算法，不允许直接调用快速排序函数，需要手动实现其逻辑过程。
def quick_sort(arr):  
    _quick_sort(arr, 0, len(arr) - 1)  
    return arr  

def _quick_sort(arr, low, high):  
    if low < high:  
        pi = _partition(arr, low, high)  
        _quick_sort(arr, low, pi - 1)  
        _quick_sort(arr, pi + 1, high)  

def _partition(arr, low, high):  
    pivot = arr[high]  
    i = low - 1  
    for j in range(low, high):  
        if arr[j] < pivot:  
            i += 1  
            arr[i], arr[j] = arr[j], arr[i]  
    arr[i + 1], arr[high] = arr[high], arr[i + 1]  
    return i + 1  

s=input("请输入一组以空格分隔的整数：")
arr=[int(item) for item in s.split()]
print("输入数组：",arr)
print("快速排序之后的数组：",insertion_sort(arr))
```

```python
#5.使用 Python 随机生成三组长度递增的随机数列，长度分别为100、1000、10000，对生成的随机数组使用不同的排序算法，例如选择排序和归并排序，至少三种排序方法。对这些数列的数据排序，并分析不同排序算法在不同长度数列下的运行效果，例如运行时间等指标。
import time
import random
import matplotlib.pyplot as plt

# 选择排序函数
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# 归并排序函数
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr

# 生成随机数列并测试排序算法
for length in [100, 1000, 10000]:
    random_list = [random.randint(1, 10000) for _ in range(length)]

    # 选择排序
    start_time = time.time()
    selection_sorted_list = selection_sort(random_list.copy())
    end_time = time.time()
    print(f"选择排序 - 长度为{length}的数列运行时间: {end_time - start_time:.6f} 秒")

    # 归并排序
    start_time = time.time()
    merge_sorted_list = merge_sort(random_list.copy())
    end_time = time.time()
    print(f"归并排序 - 长度为{length}的数列运行时间: {end_time - start_time:.6f} 秒")

    # 使用Python内置的sort方法排序
    start_time = time.time()
    built_in_sorted_list = sorted(random_list.copy())
    end_time = time.time()
    print(f"内置sort方法 - 长度为{length}的数列运行时间: {end_time - start_time:.6f} 秒")

# 存储不同长度和算法的运行时间
lengths = [100, 1000, 10000]
selection_sort_times = []
merge_sort_times = []
builtin_sort_times = []

for length in lengths:
    random_list = [random.randint(1, 10000) for _ in range(length)]

    # 选择排序时间
    start_time = time.time()
    selection_sort(random_list.copy())
    end_time = time.time()
    selection_sort_times.append(end_time - start_time)

    # 归并排序时间
    start_time = time.time()
    merge_sort(random_list.copy())
    end_time = time.time()
    merge_sort_times.append(end_time - start_time)

    # 内置sort方法时间
    start_time = time.time()
    sorted(random_list.copy())
    end_time = time.time()
    builtin_sort_times.append(end_time - start_time)

# 绘制图像
plt.plot(lengths, selection_sort_times, label='Selection Sort', marker='o')
plt.plot(lengths, merge_sort_times, label='Merge Sort', marker='s')
plt.plot(lengths, builtin_sort_times, label='Built-in Sort', marker='^')

plt.xlabel('Length of List')
plt.ylabel('Running Time (seconds)')
plt.title('Sorting Algorithm Performance Comparison')
plt.legend()
plt.show()
```
